```python
import re

tatoe_en = open("train.src", "r")
tatoe_ko = open("train.trg", "r")

out_en = open("CleanTrain.en", "w")
out_ko = open("CleanTrain.ko", "w")

end_punctuations = [ '.', '?', '!' ]

hangul = re.compile('[\u3131-\u3163\uac00-\ud7a3]+')
eng = re.compile('[A-Za-z]+')
html = re.compile(r'<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')
url = re.compile(r"https?://\S+|www\.\S+")
emoji = re.compile(r"["
                       u"\U0001F600-\U0001F64F"  # emoticons
                       u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                       u"\U0001F680-\U0001F6FF"  # transport & map symbols
                       u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                       u"\U00002702-\U000027B0"
                       u"\U000024C2-\U0001F251"
                       u"\U0001f926-\U0001f937"
                       u"\u200d"
                       u"\u2640-\u2642" 
                       "]+", flags=re.UNICODE)

emojKO = re.compile(r"["
                       u"\U0001F600-\U0001F64F"  # emoticons
                       u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                       u"\U0001F680-\U0001F6FF"  # transport & map symbols
                       u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                       u"\U00002702-\U000027B0"
                       #u"\U000024C2-\U0001F251"
                       u"\U0001f926-\U0001f937"
                       u"\u200d"
                       u"\u2640-\u2642" 
                       "]+", flags=re.UNICODE)

asciicode = re.compile(r'[^\x00-\x7f]')
#punctuations = re.compile('[@_#$%^&*()-\[\]""<>/\|}{~:;]')
punctuations = re.compile('[@_#$%^&*()\[\]“<>/\|}{~:;]')
plus = re.compile('[\+]')

#    hanCount = len(re.findall(u'[\u3130-\u318F\uAC00-\uD7A3]+', encText))

count = 1

for lineEn, lineKo in zip(tatoe_en, tatoe_ko):
    lineEn = lineEn.replace('\u200b', '').strip()
    lineEn = lineEn.replace('\u200e', '').strip()
    lineEn = lineEn.replace('\u200d', '').strip()
    lineEn = lineEn.replace('\u200c', '').strip()
    lineEn = lineEn.replace('-', '').strip()

    lineKo = lineKo.replace('\u200b', '').strip()
    lineKo = lineKo.replace('\u200c', '').strip()
    lineKo = lineKo.replace('\u200d', '').strip()
    lineKo = lineKo.replace('\u200e', '').strip()
    lineKo = lineKo.replace('-', '').strip()

    result =  [punc for punc in end_punctuations if lineEn[-1]==punc]
    if len(result)==0:
        continue

    result = re.findall(" sex ", lineEn)
    if len(result)!=0:
        continue
    result = re.findall(" sex ", lineKo)
    if len(result)!=0:
        continue
    result = re.findall("섹스", lineKo)
    if len(result)!=0:
        continue

    result = hangul.findall(lineEn)
    if len(result)!=0:
        continue

    result = html.findall(lineEn)
    if len(result)!=0:
        continue

    result = url.findall(lineEn)
    if len(result)!=0:
        continue

    result = emoji.findall(lineEn)
    if len(result)!=0:
        continue

    result = asciicode.findall(lineEn)
    if len(result)!=0:
        continue

    result = punctuations.findall(lineEn)
    if len(result)!=0:
        continue
        

    result = plus.findall(lineEn)
    if len(result)!=0:
        continue

    #############
    
    result = eng.findall(lineKo)
    if len(result)!=0:
        continue

    result = html.findall(lineKo)
    if len(result)!=0:
        continue

    result = url.findall(lineKo)
    if len(result)!=0:
        continue

    result = emojKO.findall(lineKo)
    if len(result)!=0:
        print("emojKO ", result)
        continue

    result = punctuations.findall(lineKo)
    if len(result)!=0:
        continue

    result = plus.findall(lineKo)
    if len(result)!=0:
        continue
    '''
    count = count+1
    print(count)
    if count==100:
        break
    '''
    
    if len(lineKo.split()) <= 5:
        continue
    elif len(lineKo.split()) >= 15:
        continue

    if len(lineEn.split()) <= 5:
        continue
    elif len(lineEn.split()) >= 15:
        continue

    out_en.write(lineEn+"\n")
    out_ko.write(lineKo+"\n")

    
tatoe_en.close()
tatoe_ko.close()
    
out_en.close()
out_ko.close()
```
