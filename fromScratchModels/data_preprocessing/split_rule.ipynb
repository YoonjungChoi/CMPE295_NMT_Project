Split set 95% 5% 5%

```python
fko = open("total/YJSUM.ko", "r")
fen = open("total/YJSUM.en", "r")

trainko = open("train.ko", "w")
validko = open("valid.ko", "w")
testko = open("test.ko", "w")

trainen = open("train.en", "w")
validen = open("valid.en", "w")
testen = open("test.en", "w")

line = 1

for lKo, lEn in zip(fko, fen):    
    if line%20 == 19 :
        validko.write(lKo.strip()+"\n")
        validen.write(lEn.strip()+"\n")
    elif line%20 == 0 :
        testko.write(lKo.strip()+"\n")
        testen.write(lEn.strip()+"\n")
    else:
        trainko.write(lKo.strip()+"\n")
        trainen.write(lEn.strip()+"\n")
        
    line += 1
    
fko.close()
fen.close()
trainko.close()
validko.close()
testko.close()
trainen.close()
validen.close()
testen.close()
```
