```python
import pandas as pd
import re
```

```python
def clean_str_netflix(text):
    text = str(text)
    if ('♪' in text):
        return ""
    
    text = text.replace('\u200e', '').strip()
    pattern = '-'
    text = re.sub(pattern=pattern, repl='', string=text).strip()
    pattern = "\s*\[.*?\]\s*"
    text = re.sub(pattern=pattern, repl='', string=text).strip()
    
    pattern = "\s*\(.*?\)\s*"
    text = re.sub(pattern=pattern, repl='', string=text).strip()
    return text
```

## TITLE: Emily

```python
nf_emilyinparis = pd.read_excel('netflix/download/Netflix 13621L Emily in Paris.xlsx', 
                            index_col=None, header=None)
nf_emilyinparis.info()
```

```python
nf_emilyinparis.head(20)
```

```python
out_en = open("nf_emilyinparis.en", "w")
out_ko = open("nf_emilyinparis.ko", "w")

end_punctuations = [ '.', '?', '!' ]
sentence_en = ""
sentence_ko = ""
cnt = 0
flag = False

for index, row in nf_emilyinparis.iterrows():
    cleaned_en = clean_str_netflix(row[0])
    cleaned_ko = clean_str_netflix(row[1])
    if not cleaned_en or not cleaned_ko:
        continue
        
    #print("cleaned_en:", cleaned_en)
    flag =  [punc for punc in end_punctuations if cleaned_en[-1] ==punc]
    #print(flag)
    #print(len(flag))
    
    if len(flag)==0:
        sentence_en += cleaned_en + " "
        sentence_ko += cleaned_ko + " "
    else:
        sentence_en += cleaned_en + " "
        sentence_ko += cleaned_ko + " "
        out_en.write(sentence_en.strip()+"\n")
        out_ko.write(sentence_ko.strip()+"\n")
        sentence_en = ""
        sentence_ko = ""
    
out_en.close()
out_ko.close()
```
